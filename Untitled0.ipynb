```python

import os
os.environ["XLA_PYTHON_CLIENT_PREALLOCATE"] = "false"  # đỡ ôm gần hết VRAM

import math
import numpy as np
import matplotlib.pyplot as plt

import jax
import jax.numpy as jnp

# Chọn precision ở đây:
# REAL = jnp.float32   # nhanh hơn, hợp để test
REAL = jnp.float64     # chắc hơn cho reference / convergence

jax.config.update("jax_enable_x64", REAL == jnp.float64)

print("JAX version:", jax.__version__)
print("Devices:", jax.devices())
print("Backend:", jax.default_backend())
```

```python
# ===== Cell 2: output folders =====
BASE_DIR = "/content/burgers2d_results"
IMG_DIR = os.path.join(BASE_DIR, "images")
DATA_DIR = os.path.join(BASE_DIR, "data")

os.makedirs(IMG_DIR, exist_ok=True)
os.makedirs(DATA_DIR, exist_ok=True)

print("Saving to:", BASE_DIR)
```

```python
# ===== Cell 3: grid + spectral operators =====
def make_grid(nx, ny, Lx=2.0 * jnp.pi, Ly=2.0 * jnp.pi):
    x = jnp.linspace(0.0, Lx, nx, endpoint=False)
    y = jnp.linspace(0.0, Ly, ny, endpoint=False)
    X, Y = jnp.meshgrid(x, y, indexing="ij")

    dx = Lx / nx
    dy = Ly / ny

    kx = 2.0 * jnp.pi * jnp.fft.fftfreq(nx, d=float(dx))
    ky = 2.0 * jnp.pi * jnp.fft.fftfreq(ny, d=float(dy))
    KX, KY = jnp.meshgrid(kx, ky, indexing="ij")
    K2 = KX**2 + KY**2

    # 2/3 dealiasing mask
    kx_max = jnp.max(jnp.abs(kx))
    ky_max = jnp.max(jnp.abs(ky))
    mask = (
        (jnp.abs(KX) <= (2.0 / 3.0) * kx_max) &
        (jnp.abs(KY) <= (2.0 / 3.0) * ky_max)
    ).astype(jnp.float64)

    return x, y, X, Y, dx, dy, KX, KY, K2, mask


def spectral_filter(f, mask):
    fhat = jnp.fft.fft2(f) * mask
    return jnp.fft.ifft2(fhat).real


def spectral_grad(f, KX, KY, mask):
    fhat = jnp.fft.fft2(f) * mask
    fx = jnp.fft.ifft2(1j * KX * fhat).real
    fy = jnp.fft.ifft2(1j * KY * fhat).real
    return fx, fy


def diffuse_half_step(f, Ehalf, mask):
    fhat = jnp.fft.fft2(f) * mask
    return jnp.fft.ifft2(Ehalf * fhat).real
```

```python
# ===== Cell 4: advective RHS + time stepper =====
def adv_rhs(u, v, KX, KY, mask):
    u = spectral_filter(u, mask)
    v = spectral_filter(v, mask)

    ux, uy = spectral_grad(u, KX, KY, mask)
    vx, vy = spectral_grad(v, KX, KY, mask)

    du = -(u * ux + v * uy)
    dv = -(u * vx + v * vy)
    return du, dv


def rk4_advective(u, v, dt, KX, KY, mask):
    k1u, k1v = adv_rhs(u, v, KX, KY, mask)

    u2 = u + 0.5 * dt * k1u
    v2 = v + 0.5 * dt * k1v
    k2u, k2v = adv_rhs(u2, v2, KX, KY, mask)

    u3 = u + 0.5 * dt * k2u
    v3 = v + 0.5 * dt * k2v
    k3u, k3v = adv_rhs(u3, v3, KX, KY, mask)

    u4 = u + dt * k3u
    v4 = v + dt * k3v
    k4u, k4v = adv_rhs(u4, v4, KX, KY, mask)

    un = u + (dt / 6.0) * (k1u + 2.0 * k2u + 2.0 * k3u + k4u)
    vn = v + (dt / 6.0) * (k1v + 2.0 * k2v + 2.0 * k3v + k4v)

    un = spectral_filter(un, mask)
    vn = spectral_filter(vn, mask)
    return un, vn


def make_spectral_smoother(K2, strength=24.0, order=8):
    k2max = jnp.max(K2)
    sigma = jnp.exp(-strength * (K2 / k2max) ** (order / 2))
    return sigma

def spectral_filter_smooth(f, mask, sigma):
    fhat = jnp.fft.fft2(f) * mask
    return jnp.fft.ifft2(fhat * sigma).real

def build_stepper(KX, KY, K2, mask, nu, dt, eps4=0.0):
    # diffusion + optional hyperviscosity
    Ehalf = jnp.exp(-0.5 * dt * (nu * K2 + eps4 * K2**2))
    sigma = make_spectral_smoother(K2, strength=24.0, order=8)

    @jax.jit
    def step(u, v):
        u = diffuse_half_step(u, Ehalf, mask)
        v = diffuse_half_step(v, Ehalf, mask)

        u, v = rk4_advective(u, v, dt, KX, KY, mask)

        u = diffuse_half_step(u, Ehalf, mask)
        v = diffuse_half_step(v, Ehalf, mask)

        u = spectral_filter_smooth(u, mask, sigma)
        v = spectral_filter_smooth(v, mask, sigma)
        return u, v

    return step


def recommend_dt(u0, v0, dx, dy, nu, cfl=0.12):
    umax = max(float(jnp.max(jnp.abs(u0))), 1e-8)
    vmax = max(float(jnp.max(jnp.abs(v0))), 1e-8)

    adv_dt = cfl * min(float(dx) / umax, float(dy) / vmax)
    diff_dt = 0.25 * min(float(dx) ** 2, float(dy) ** 2) / max(nu, 1e-12)

    return min(adv_dt, diff_dt)
```

```python
# ===== Cell 5: initial condition =====
def init_condition(X, Y):
    # Smooth but capable of steepening for small viscosity
    u0 = 1.5 * (
        jnp.sin(X)
        + 0.50 * jnp.sin(2.0 * X)
        + 0.25 * jnp.sin(X + Y)
    )
    v0 = 1.5 * (
        jnp.sin(Y)
        + 0.50 * jnp.sin(2.0 * Y)
        - 0.25 * jnp.sin(X - Y)
    )
    return u0, v0
```

```python
# ===== Cell 6: simulation =====
def simulate(u0, v0, stepper, nsteps, save_every=10):
    u, v = u0, v0
    us = [jax.device_get(u0)]
    vs = [jax.device_get(v0)]

    for n in range(nsteps):
        u, v = stepper(u, v)
        if (n + 1) % save_every == 0:
            us.append(jax.device_get(u))
            vs.append(jax.device_get(v))

    return np.stack(us), np.stack(vs)
```

```python
# ===== fast final-state simulation =====
def simulate_final(u0, v0, stepper, nsteps):
    u, v = u0, v0
    for _ in range(nsteps):
        u, v = stepper(u, v)
    return jax.device_get(u), jax.device_get(v)
```

```python
# ===== Cell 7: plotting =====
def save_fields_plot(x, y, u, v, grad_u, out_path, title_prefix=""):
    fig, axes = plt.subplots(1, 3, figsize=(16, 4.5))

    extent = [float(x[0]), float(x[-1] + (x[1] - x[0])),
              float(y[0]), float(y[-1] + (y[1] - y[0]))]

    im0 = axes[0].imshow(np.array(u).T, origin="lower", extent=extent, aspect="auto")
    axes[0].set_title(f"{title_prefix} u(x,y,T)")
    axes[0].set_xlabel("x")
    axes[0].set_ylabel("y")
    plt.colorbar(im0, ax=axes[0])

    im1 = axes[1].imshow(np.array(v).T, origin="lower", extent=extent, aspect="auto")
    axes[1].set_title(f"{title_prefix} v(x,y,T)")
    axes[1].set_xlabel("x")
    axes[1].set_ylabel("y")
    plt.colorbar(im1, ax=axes[1])

    im2 = axes[2].imshow(np.array(grad_u).T, origin="lower", extent=extent, aspect="auto")
    axes[2].set_title(f"{title_prefix} |grad u|(x,y,T)")
    axes[2].set_xlabel("x")
    axes[2].set_ylabel("y")
    plt.colorbar(im2, ax=axes[2])

    plt.tight_layout()
    plt.savefig(out_path, dpi=160, bbox_inches="tight")
    plt.show()
    plt.close(fig)
```

```python
# ===== Cell 9A: viscosity sweep at fixed resolution =====
nus = [1e-2, 1e-3, 5e-4]
N = 256
T_map = {
    1e-2: 0.4,
    1e-3: 0.3,
    5e-4: 0.25,
}
cfl_map = {
    1e-2: 0.02,
    1e-3: 0.01,
    5e-4: 0.005,
}
eps4_map = {
    1e-2: 1e-5,
    1e-3: 5e-5,
    5e-4: 1e-4,
}

summary = []

nx = ny = N
x, y, X, Y, dx, dy, KX, KY, K2, mask = make_grid(nx, ny)
u0, v0 = init_condition(X, Y)

for nu in nus:
    T = T_map[nu]
    cfl = cfl_map[nu]
    eps4 = eps4_map[nu]

    print(f"\nRunning viscosity case: N={N}, nu={nu}, T={T}, cfl={cfl}, eps4={eps4}")

    dt = recommend_dt(u0, v0, dx, dy, nu, cfl=cfl)
    nsteps = int(np.ceil(T / dt))
    dt = T / nsteps

    stepper = build_stepper(KX, KY, K2, mask, nu, dt, eps4=eps4)
    save_every = max(1, nsteps // 40)

    us, vs = simulate(u0, v0, stepper, nsteps, save_every=save_every)

    uT = jnp.array(us[-1])
    vT = jnp.array(vs[-1])

    ux, uy = spectral_grad(uT, KX, KY, mask)
    grad_u = jnp.sqrt(ux**2 + uy**2)

    case_name = f"burgers2d_visc_N{N}_nu{nu:.0e}"

    img_path = os.path.join(IMG_DIR, case_name + ".png")
    save_fields_plot(
        x, y, uT, vT, grad_u,
        out_path=img_path,
        title_prefix=f"nu={nu}, N={N}"
    )

    data_path = os.path.join(DATA_DIR, case_name + ".npz")
    np.savez(
        data_path,
        x=np.array(x),
        y=np.array(y),
        us=us,
        vs=vs,
        nu=nu,
        T=T,
        dt=dt,
        nsteps=nsteps,
        save_every=save_every,
        N=N,
        cfl=cfl,
        eps4=eps4,
    )

    E0 = 0.5 * float(jnp.sum(jnp.array(us[0])**2 + jnp.array(vs[0])**2) * dx * dy)
    ET = 0.5 * float(jnp.sum(uT**2 + vT**2) * dx * dy)

    summary.append({
        "case": case_name,
        "mode": "viscosity_sweep",
        "N": N,
        "nu": nu,
        "T": T,
        "cfl": cfl,
        "eps4": eps4,
        "dt": dt,
        "nsteps": nsteps,
        "saved_frames": len(us),
        "E0": E0,
        "ET": ET,
        "max_grad_u": float(jnp.max(grad_u)),
        "u_min": float(jnp.min(uT)),
        "u_max": float(jnp.max(uT)),
        "v_min": float(jnp.min(vT)),
        "v_max": float(jnp.max(vT)),
    })

print("\nViscosity sweep done.")
```

```python
# ===== Cell 9B: grid sweep at fixed viscosity =====
resolutions = [128, 256]
nu = 1e-2
T = 0.4
cfl = 0.02
eps4 = 1e-5

for N in resolutions:
    nx = ny = N
    x, y, X, Y, dx, dy, KX, KY, K2, mask = make_grid(nx, ny)
    u0, v0 = init_condition(X, Y)

    print(f"\nRunning grid case: N={N}, nu={nu}, T={T}, cfl={cfl}, eps4={eps4}")

    dt = recommend_dt(u0, v0, dx, dy, nu, cfl=cfl)
    nsteps = int(np.ceil(T / dt))
    dt = T / nsteps

    stepper = build_stepper(KX, KY, K2, mask, nu, dt, eps4=eps4)
    save_every = max(1, nsteps // 40)

    us, vs = simulate(u0, v0, stepper, nsteps, save_every=save_every)

    uT = jnp.array(us[-1])
    vT = jnp.array(vs[-1])

    ux, uy = spectral_grad(uT, KX, KY, mask)
    grad_u = jnp.sqrt(ux**2 + uy**2)

    case_name = f"burgers2d_grid_N{N}_nu{nu:.0e}"

    img_path = os.path.join(IMG_DIR, case_name + ".png")
    save_fields_plot(
        x, y, uT, vT, grad_u,
        out_path=img_path,
        title_prefix=f"nu={nu}, N={N}"
    )

    data_path = os.path.join(DATA_DIR, case_name + ".npz")
    np.savez(
        data_path,
        x=np.array(x),
        y=np.array(y),
        us=us,
        vs=vs,
        nu=nu,
        T=T,
        dt=dt,
        nsteps=nsteps,
        save_every=save_every,
        N=N,
        cfl=cfl,
        eps4=eps4,
    )

    E0 = 0.5 * float(jnp.sum(jnp.array(us[0])**2 + jnp.array(vs[0])**2) * dx * dy)
    ET = 0.5 * float(jnp.sum(uT**2 + vT**2) * dx * dy)

    summary.append({
        "case": case_name,
        "mode": "grid_sweep",
        "N": N,
        "nu": nu,
        "T": T,
        "cfl": cfl,
        "eps4": eps4,
        "dt": dt,
        "nsteps": nsteps,
        "saved_frames": len(us),
        "E0": E0,
        "ET": ET,
        "max_grad_u": float(jnp.max(grad_u)),
        "u_min": float(jnp.min(uT)),
        "u_max": float(jnp.max(uT)),
        "v_min": float(jnp.min(vT)),
        "v_max": float(jnp.max(vT)),
    })

print("\nGrid sweep done.")
```

```python
# ===== Reference utilities =====
def run_case(N, nu, T, cfl, eps4, case_prefix, save_frames=False):
    nx = ny = N
    x, y, X, Y, dx, dy, KX, KY, K2, mask = make_grid(nx, ny)
    u0, v0 = init_condition(X, Y)

    dt = recommend_dt(u0, v0, dx, dy, nu, cfl=cfl)
    nsteps = int(np.ceil(T / dt))
    dt = T / nsteps

    stepper = build_stepper(KX, KY, K2, mask, nu, dt, eps4=eps4)

    if save_frames:
        save_every = max(1, nsteps // 40)
        us, vs = simulate(u0, v0, stepper, nsteps, save_every=save_every)
        uT = us[-1]
        vT = vs[-1]
    else:
        uT, vT = simulate_final(u0, v0, stepper, nsteps)
        us, vs = None, None
        save_every = None

    data_path = os.path.join(DATA_DIR, f"{case_prefix}.npz")
    np.savez(
        data_path,
        x=np.array(x),
        y=np.array(y),
        uT=uT,
        vT=vT,
        nu=nu,
        T=T,
        dt=dt,
        nsteps=nsteps,
        N=N,
        cfl=cfl,
        eps4=eps4,
        save_every=(-1 if save_every is None else save_every),
        us=(np.array([]) if us is None else us),
        vs=(np.array([]) if vs is None else vs),
    )

    return {
        "path": data_path,
        "x": np.array(x),
        "y": np.array(y),
        "uT": uT,
        "vT": vT,
        "nu": nu,
        "T": T,
        "dt": dt,
        "nsteps": nsteps,
        "N": N,
        "cfl": cfl,
        "eps4": eps4,
    }
```

```python
def relative_l2(a, b):
    return np.linalg.norm(a - b) / max(np.linalg.norm(b), 1e-14)

def compare_with_reference(test_case, ref_case):
    N_test = test_case["N"]
    N_ref = ref_case["N"]

    if N_ref % N_test != 0:
        raise ValueError("Reference resolution must be an integer multiple of test resolution.")

    stride = N_ref // N_test

    u_ref_ds = ref_case["uT"][::stride, ::stride]
    v_ref_ds = ref_case["vT"][::stride, ::stride]

    u_test = test_case["uT"]
    v_test = test_case["vT"]

    err_u = relative_l2(u_test, u_ref_ds)
    err_v = relative_l2(v_test, v_ref_ds)

    return {
        "stride": stride,
        "u_ref_ds": u_ref_ds,
        "v_ref_ds": v_ref_ds,
        "err_u_rel_l2": err_u,
        "err_v_rel_l2": err_v,
    }
```

```python
# ===== Run one test-reference pair =====
nu = 1e-2
T = 0.4

test_case = run_case(
    N=256,
    nu=nu,
    T=T,
    cfl=0.02,
    eps4=1e-5,
    case_prefix="burgers2d_test_N256_nu1e-02",
    save_frames=False,
)

ref_case = run_case(
    N=512,  # Changed from 384 to 256 to avoid N > 256 performance issues and satisfy N_ref % N_test == 0
    nu=nu,
    T=T,
    cfl=0.005,
    eps4=1e-5,
    case_prefix="burgers2d_ref_N256_nu1e-02", # Updated case name
    save_frames=False,
)

cmp = compare_with_reference(test_case, ref_case)

print("Test path:", test_case["path"])
print("Reference path:", ref_case["path"])
print("stride =", cmp["stride"])
print("Relative L2 error for u =", cmp["err_u_rel_l2"])
print("Relative L2 error for v =", cmp["err_v_rel_l2"])
```

### 10. Spatial Convergence Study

We will conduct a spatial convergence study by running simulations at different resolutions and comparing them to a higher-resolution reference solution. This helps us understand the convergence speed of our digital solver.

```python
spatial_convergence_results = []

nu_convergence = 1e-2
T_convergence = 0.4
cfl_test_convergence = 0.02
eps4_convergence = 1e-5

test_resolutions = [64, 128, 256]

ref_resolution = 512
cfl_ref_convergence = 0.005

print(f"Running high-resolution reference case (N={ref_resolution}, nu={nu_convergence})...")
ref_case_convergence = run_case(
    N=ref_resolution,
    nu=nu_convergence,
    T=T_convergence,
    cfl=cfl_ref_convergence,
    eps4=eps4_convergence,
    case_prefix=f"burgers2d_ref_N{ref_resolution}_nu{nu_convergence:.0e}_convergence",
    save_frames=False,
)
print("Reference case done.")

for N_test in test_resolutions:
    print(f"\nRunning test case (N={N_test}, nu={nu_convergence})...")
    test_case_convergence = run_case(
        N=N_test,
        nu=nu_convergence,
        T=T_convergence,
        cfl=cfl_test_convergence,
        eps4=eps4_convergence,
        case_prefix=f"burgers2d_test_N{N_test}_nu{nu_convergence:.0e}_convergence",
        save_frames=False,
    )

    # So sánh trường hợp thử nghiệm với trường hợp tham chiếu
    cmp_convergence = compare_with_reference(test_case_convergence, ref_case_convergence)

    spatial_convergence_results.append({
        "N_test": N_test,
        "N_ref": ref_resolution,
        "nu": nu_convergence,
        "T": T_convergence,
        "cfl_test": cfl_test_convergence,
        "cfl_ref": cfl_ref_convergence,
        "err_u_rel_l2": cmp_convergence["err_u_rel_l2"],
        "err_v_rel_l2": cmp_convergence["err_v_rel_l2"],
    })

convergence_df = pd.DataFrame(spatial_convergence_results)
print("\nSpatial convergence study completed.")
display(convergence_df)
```
